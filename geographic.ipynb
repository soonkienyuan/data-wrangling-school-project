```python
from bokeh.plotting import figure
from bokeh.tile_providers import get_provider, WIKIMEDIA
from bokeh.io import output_notebook, show, curdoc
from bokeh.models import CustomJS, ColumnDataSource, CheckboxButtonGroup, RangeSlider
from bokeh.layouts import column, row
output_notebook()
```

```python
from pyproj import Proj, transform
# cartography(map making field), all maps are projected
# then we use Mercator projection here
inProj = Proj(init='epsg:3857')
outProj = Proj(init='epsg:4326')

wikimedia = get_provider(WIKIMEDIA)

aus_lon1, aus_lat1 = transform(outProj,inProj, df['Longitude'].min(), df['Latitude'].min())
aus_lon2, aus_lat2 = transform(outProj,inProj,df['Longitude'].max(), df['Latitude'].max())
temp_df = df
df_src = ColumnDataSource(temp_df)

p = figure(plot_width=900, plot_height=700,
           x_range=(aus_lon1, aus_lon2), y_range=(aus_lat1, aus_lat2),
           x_axis_type="mercator", y_axis_type="mercator",
           tooltips=[("(Long, Lat)", "(@Longitude, @Latitude)")],
           title="Austin Bicycle Crashes")
p.add_tile(wikimedia)

circle_glyph = p.circle(x="MercatorX", y="MercatorY",
         size=7,
         fill_color="red", line_color="red",
         fill_alpha=0.9,source=df_src)


severity_picker = CheckboxButtonGroup(labels=severity_types)
Injury_Count_Picker = RangeSlider(start=0, end=df['Crash Incapacitating Injury Count'].max(), value=(1,2), step=1, title="Stuff")

def severity_picker_handler(attr):
    activated = list(severity_picker.active)
    if len(activated) > 0:
        activated = [severity_types[i] for i in activated]
        print(activated)
        print(temp_df[temp_df['Crash Severity'].isin(activated)])
        p.title.text = f"{activated} Austin Cyclist"
        df_src.data = temp_df[temp_df['Crash Severity'].isin(activated)]

def injury_picker_handler(attr, old, new):
    new_start, new_end = Injury_Count_Picker.value
    circle_glyph.glyph.fill_color = getRandomColor()
    df_src.data = temp_df[temp_df['Crash Incapacitating Injury Count'].between(new_start,new_end)]
    # circle_glyph.data_source = ColumnDataSource(temp_src)

severity_picker.on_click(severity_picker_handler)
Injury_Count_Picker.on_change('value', injury_picker_handler)

layout = column(severity_picker, Injury_Count_Picker, p)

curdoc().add_root(layout)
```
